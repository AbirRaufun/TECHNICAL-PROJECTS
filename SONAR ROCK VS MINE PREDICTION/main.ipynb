```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
import matplotlib.pyplot as plt
```

```python
sonar_data=pd.read_csv('G:\AIUB\8th\ML notes\projects\sonar\sonar.csv', header=None)
```

```python
X=sonar_data.drop(columns=60,axis=1)
Y=sonar_data[60]
```

```python
print(X)
print(Y)
```

```python
x_train,x_text,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,x_train.shape,x_text.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pre=model.predict(x_train)
tranning_data_ac=accuracy_score(x_train_pre,y_train)
```

```python
print("Ac of trainning : ",tranning_data_ac)
```

```python
x_text_pre=model.predict(x_text)
test_data_ac=accuracy_score(x_text_pre,y_test)
```

```python
print("Ac of testing : ",test_data_ac)
```

```python
cm = confusion_matrix(y_test, x_text_pre, labels=model.classes_)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```
